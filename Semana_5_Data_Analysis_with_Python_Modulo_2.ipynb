```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
```

```python
#Definir nombres de variables a utilizar para descargar el archivo de base de datos
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
#Descargar archivo y mostrar los primeros 5 registros
df = pd.read_csv(path, names = headers)
df.head()
```

```python
#Identify and handling missing data
```

```python
#Reemplazar "?" con NaN y almacenarlos en el mismo dataframe
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Evaluating for missing data
```

```python
#Identificar si en dataframe se encuentran datos nulos y mostrar el resultado de las primeras 5 columnas
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Como se encontraron algunos resultados con True nos indica que si hay datos nulos en nuestro data frame
```

```python
#Count missing values in each column
```

```python
#Identificar en cada columna cuantos valores nulos hay utilizando la funcion for
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
#Dealing with missing data
```

```python
#Calcular la media para la columna "normalized-losses"
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#Reemplazar los valores Nan de la columna "normalized-losses" con el valor de la media obtenido
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
df.head()
```

```python
#Calcular la media para la columna "bore"
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
#Reemplazar los valores Nan de la columna "bore" con el valor de la media obtenido
df["bore"].replace(np.nan, avg_bore, inplace=True)
df.head()
```

```python
#Calcular la media para la columna "stroke"
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
#Reemplazar los valores Nan de la columna "stroke" con el valor de la media obtenido
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df.head()
```

```python
#Calcular la media para la columna "horsepower"
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
#Reemplazar los valores Nan de la columna "horsepower" con el valor de la media obtenido
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
df.head()
```

```python
#Calcular la media para la columna "peak-rpm"
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
#Reemplazar los valores Nan de la columna "peak-rpm" con el valor de la media obtenido
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
df.head()
```

```python
#Otro metodo para reemplazar valores tipo object, identificar primero cuantos valores hay en una columna "num-of-doors"
df['num-of-doors'].value_counts()
```

```python
#Identificar de los valores en la columna "num-of-doors" cual es el que tiene la mayor cantidad
df['num-of-doors'].value_counts().idxmax()
```

```python
#Reemplazar valores NaN de la columna "num-of-doors" con el valor obtenido
df["num-of-doors"].replace(np.nan, "four", inplace=True)
df.head()
```

```python
#Eliminar filas de la columna "price" con valor NaN
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
#Correct data format
```

```python
#Identificar los data types de cada columna
df.dtypes
```

```python
#Se identifican columnas con datos tipo object cuando deberian ser int o float, se procede a corregir los data types
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#Identificar nuevamente los data types de cada columna
df.dtypes
```

```python
#Data Stardization
```

```python
#Transformar de la columan "city-mpg" de MPG a L/100km en una nueva columna
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
#Repetir el mismo procedimiento para la columna "highway-mpg" y reemplazar la misma columna por "highway-L/100km"
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

```python
#Data Normalization
```

```python
#Normalizar las columnas "lenght" y "width" usando el valor maximo
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df.head()
```

```python
#Realizar el mismo procedimiento para la columna "height" y mostrar los valores para las primeras 5 filas de las 3 columnas modificadas
df['height'] = df['height']/df['height'].max()
df[["length","width","height"]].head()
```

```python
#Binning
```

```python
#Convertir datos al formato correcto
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
#Graficar "horsepower"
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Se crean 3 bins del mismo tamao realizando 4 divisiones
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
#Damos nombres a los bins y los asignamos en df
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Se revisa cuantas repeticiones hay por cada bin
df["horsepower-binned"].value_counts()
```

```python
#Se vuelve a graficar utilizando los bins
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Bins Visualization
```

```python
#Se grafica en forma de histograma
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
#Se revisan las columnas
df.columns
```

```python
#Para la columna "fuel-type" se crean indicadores de las variables y se asignan a variable dummy
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
#Se anexan columnas de "diesel" y "gas" a df y se elimina columna "fuel-type"
df = pd.concat([df, dummy_variable_1], axis=1)
df.drop("fuel-type", axis = 1, inplace=True)
df.head()
```

```python
#Para la columna "aspiration" se crean indicadores de las variables y se asignan a variable dummy
dummy_variable_2 = pd.get_dummies(df['aspiration'])
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
#Se anexan columnas de "aspiration-std" y "aspiration-turbo" a df y se elimina columna "aspiration"
df = pd.concat([df, dummy_variable_2], axis=1)
df.drop('aspiration', axis = 1, inplace=True)
df.head()
```

```python
#Se guarda dataframe con modificaciones en un .csv
df.to_csv('clean_df.csv')
```
